```python
import pandas as pd
df=pd.read_csv("/content/drive/MyDrive/Progetto ML/Student_Performance.csv")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
#transformation of the extracurricular activities column (No=0; Yes=1)
df['Extracurricular Activities']=df['Extracurricular Activities'].replace({'Yes':1, 'No':0})
```

```python
df.head()
```

```python
#normalization of the dataframe
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

scaler=Pipeline([
    ('std_scaler', StandardScaler())
])
df_scaled_array=scaler.fit_transform(df)
df_scaled=pd.DataFrame(df_scaled_array, columns=df.columns)
df_scaled.head()
```

```python
#80% training set, 20% test set

from sklearn.model_selection import train_test_split
training_set, test_set=train_test_split(df_scaled, test_size=0.2, random_state=42)
training_set=training_set.copy()
test_set=test_set.copy()
```

```python
X_train=training_set.drop('Performance Index', axis=1)
X_train.shape
```

```python
Y_train=training_set['Performance Index']
Y_train.shape
```

```python
#training
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(X_train,Y_train)
```

```python
import numpy as np
from sklearn.metrics import mean_squared_error

Y_train_hat=model.predict(X_train)
mse=mean_squared_error(Y_train, Y_train_hat)
rmse=np.sqrt(mse)
rmse
```

```python
X_test=test_set.drop('Performance Index', axis=1)
X_test.shape
```

```python
Y_test=test_set['Performance Index']
Y_test.shape
```

```python
Y_test_hat=model.predict(X_test)
mse=mean_squared_error(Y_test, Y_test_hat)
rmse=np.sqrt(mse)
rmse
```
